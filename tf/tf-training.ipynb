```python
def f(x):
    return 5 * x + 3
```

```python
def show(f):
    plt.scatter(data_x, data_y)
    plt.plot(data_x, f(data_x))

show(f)
```

```python
a = tf.Variable(5.0)
b = tf.Variable(3.0)

def f(x):
    return a * x + b

show(f)
```

```python
def loss(predicted_y, target_y):
    return tf.reduce_mean(tf.square(predicted_y - target_y))
```

```python
loss(f(data_x), data_y)
```

```python
for i in range(30):
    with tf.GradientTape() as t:
        ls = loss(f(data_x), data_y)
    print("Loss", ls.numpy())
    ga, gb = t.gradient(ls, [a, b])
    a.assign_sub(ga * 0.0001)
    b.assign_sub(gb * 0.0001)

print(a, b)
show(f)
```
